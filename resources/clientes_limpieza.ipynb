```python
def cleaning(x):
    x['DIRECCIÓN'] = ""
    if pd.notna(x['Calle']):
        x['DIRECCIÓN'] += x['Calle']
    if pd.notna(x['Número']):
        space = ", " if x["DIRECCIÓN"] else ""
        x['DIRECCIÓN'] += space + str(x['Número'])
    
    x["LOCALIDAD"] = ""
    if pd.notna(x['Localidad']):
        x["LOCALIDAD"] = x['Localidad']
    if pd.notna(x['CP']):
        space = ", " if x["LOCALIDAD"] else ""
        x["LOCALIDAD"] += space + "CP: " + str(x['CP'])
    if pd.notna(x['Ciudad']):
        space = ", " if x["LOCALIDAD"] else ""
        x["LOCALIDAD"] += space + x['Ciudad']
    if pd.notna(x['Provincia']):
        space = ", " if x["LOCALIDAD"] else ""
        x["LOCALIDAD"] += space + x['Provincia']
    
    x['TELEFONO'] = ""
    if pd.notna(x['Tel I']):
        x['TELEFONO'] = str(x['Tel I'])
    if pd.notna(x['Tel II']):
        space = ", " if x["TELEFONO"] else ""
        x['TELEFONO'] += space + str(x['Tel II'])
    
    x['Otros datos'] = ""
    if x["Web Site"] != "http://":
        x['Otros datos'] = x["Web Site"]
    if pd.notna(x['Fuente']):
        x['Otros datos'] = str(x['Otros datos'])  # Ensure it's a string
        space = ", " if x["Otros datos"] else ""
        x['Otros datos'] += space + "Fuente: " + str(x['Fuente'])
    if pd.notna(x["Industria '()' - SubIndustria '[]'"]):
        x['Otros datos'] = str(x['Otros datos'])  # Ensure it's a string
        space = ", " if x["Otros datos"] else ""
        x['Otros datos'] += space + x["Industria '()' - SubIndustria '[]'"]
    
    return x

df_clean = df_unique.apply(cleaning, axis=1)
df_clean = df_clean.drop(columns=['Calle', 'Calificación', 'Dirección Descripción', 'Número','País', 'Localidad', 'CP', 'Ciudad', 'Provincia', 'Tel I', 'Tel II', 'Web Site', 'Fuente', "Industria '()' - SubIndustria '[]'"])
#rename empresa to CLIENTE
df_clean = df_clean.rename(columns={'Empresa': 'CLIENTE'})
df_clean.head()
```

```python
import asyncio
import aiomysql
import pandas as pd
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# MySQL database credentials from environment variables
credentials = {
    'host': os.getenv('MYSQL_DB_HOST'),
    'user': os.getenv('MYSQL_DB_USER'),
    'db': 'energia_global_ventas',
    'password': os.getenv('MYSQL_DB_PASSWORD'),
    'port': int(os.getenv('MYSQL_DB_PORT', 3306)),  # Default to 3306 if not set
}

# Rename 'CUIT' column to 'DNI/CUIT'
df_clean = df_clean.rename(columns={'CUIT': 'DNI/CUIT'})

# Replace NaN values in the DataFrame with None
df_clean = df_clean.where(pd.notnull(df_clean), None)

#511 lines were inserted, delete the first 511 lines
df_clean = df_clean.iloc[734:]

async def create_customers(pool, batch):
    async with pool.acquire() as conn:
        async with conn.cursor() as cur:
            sql = """
            INSERT INTO CLIENTES (CLIENTE, `DNI/CUIT`, `DIRECCIÓN`, LOCALIDAD, TELEFONO, `Otros datos`)
            VALUES (%s, %s, %s, %s, %s, %s)
            """
            await cur.executemany(sql, [
                (
                    row['CLIENTE'],
                    row['DNI/CUIT'],
                    row['DIRECCIÓN'],
                    row['LOCALIDAD'],
                    row['TELEFONO'],
                    row['Otros datos']
                ) for _, row in batch.iterrows()
            ])
            await conn.commit()
            print(f"{len(batch)} customers created successfully.")

async def create_customers_in_batches(df, batch_size=500):
    pool = await aiomysql.create_pool(**credentials)
    async with pool:
        for start in range(0, len(df), batch_size):
            batch = df.iloc[start:start + batch_size]
            await create_customers(pool, batch)

# Execute the async function
await create_customers_in_batches(df_clean)
```
